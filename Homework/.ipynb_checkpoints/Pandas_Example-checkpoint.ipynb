```python
import pandas as pd
```

## 1) read_csv() fonksiyonu
Dosya içeriğini okumamızı sağlar.

```python
df1 = pd.read_csv("masaustu_data_kaggle.csv")
df1
```

## 2) head() fonksiyonu 
İlk 5 satırı görüntüleme yapar 

```python
df1.head()
```

## 3) info() fonksiyonu
Veri çerçevesi hakkında genel bilgi alma yapar.

```python
df1.info()
```

## 4) describe() fonksiyonu
İstatistiksel özet çıkarma yapar.

```python
df1.describe()
```

## 5) isnull() fonksiyonu 
eksik (NaN) değerleri tespit etmek için kullanılır.

```python
df1.isnull()
```

## 6) fillna() fonksiyonu 
Eksik değerleri doldurmak için kullanılır. Doldurma işlemi tek bir değerle veya farklı değerlerle yapılabilir.

```python
df1.fillna(0)
```

## 7) sort_values() fonksiyonu
Verileri, belirli bir sütuna göre sıralamak için kullanılır.

```python
df1.sort_values(by='Temel İşlemci Hızı (GHz)', ascending=True)
```

## 8) value_counts() fonksiyonu
Bir sütundaki her bir benzersiz değerin kaç kere tekrar ettiğini sayar.

```python
df1['İşletim Sistemi'].value_counts()
```

## 9) memory_usage() fonksiyonu
Bellek kullanımını hesaplar.

```python
df1.memory_usage()
```

## 10) dropna() fonksiyonu
Eksik verileri kaldırma yapar.

```python
df1.dropna()
```
